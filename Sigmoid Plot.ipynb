# Plotting sigmoid functions

Importing required packages

```python
import numpy as np
import matplotlib.pyplot as plt
```

Creating numpy as array from (-1,1] with step of 0.1

```python
X=np.arange(-1,1,0.1)
```

### Experimentation with sigmoid function

###### Plotting individual functions

>1/(1+np.exp(- 400*X - 24))

>1/(1+np.exp(- 400*X + 24))

```python
Y11=1/(1+np.exp(- 400*X - 24))
fig=plt.figure()
plt.title("H11 Plot")
plt.plot(X, Y11, linewidth=3, color='b')
plt.xlabel('X')
plt.ylabel('Y11')
plt.grid()
plt.show()
```

```python
Y12=1/(1+np.exp(- 400*X + 24))
fig=plt.figure()
plt.title("H12 Plot")
plt.plot(X, Y12, linewidth=3, color='b')
plt.xlabel('X')
plt.ylabel('Y12')
plt.grid()
plt.show()
```

###### Difference of the two sigmoid functions

```python
Y21=Y11 - Y12
fig=plt.figure()
plt.title("H21 Plot")
plt.plot(X, Y21, linewidth=3, color='b')
plt.xlabel('X')
plt.ylabel('Y21')
plt.grid()
plt.show()
```

```python
Y22=Y12 - Y11
fig=plt.figure()
plt.title("H22 Plot")
plt.plot(X, Y22, linewidth=3, color='b')
plt.xlabel('X')
plt.ylabel('Y22')
plt.grid()
plt.show()
```

###### Sum of the given sigmoid functions

```python
Y31=Y11 + Y12
fig=plt.figure()
plt.title("H31 Plot")
plt.plot(X, Y31, linewidth=3, color='b')
plt.xlabel('X')
plt.ylabel('Y31')
plt.grid()
plt.show()
```
